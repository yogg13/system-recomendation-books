```python
ratings_with_books = ratings.merge(books, on="ISBN")
```

```python
count_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
count_rating.rename(columns={'rating': 'book_rating'}, inplace=True)
```

```python
rating_final = ratings_with_books.merge(count_rating, on='title')
```

```python
rating_final = rating_final[rating_final['book_rating']>=50]
```

```python
rating_final.sample(15)
```

```python
rating_final.shape
```

```python
rating_final.drop_duplicates(['title', 'user_id'], inplace=True)
```

```python
rating_final.shape
```

```python
book_pivot = rating_final.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[237, :].values.reshape(1, -1), n_neighbors=6)
```

```python
import pickle
pickle.dump(model, open('artifacts/model.pkl', 'wb'))
pickle.dump(book_name, open('artifacts/book_name.pkl', 'wb'))
pickle.dump(rating_final, open('artifacts/rating_final.pkl', 'wb'))
pickle.dump(book_pivot, open('artifacts/book_pivot.pkl', 'wb'))
```

```python
def book_recomendation(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6)

    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            print(j)
```
